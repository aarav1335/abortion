## Abortions By State

```python
import pandas as pd

df = pd.read_excel("GuttmacherInstituteAbortionDataByState.xlsx")
df.head()
```

## 2016-2020 Presidential Election Results

```python
democrat_states = [
    "California", "Colorado", "Connecticut", "Delaware", "Hawaii",
    "Illinois", "Maine", "Maryland", "Massachusetts", "Minnesota",
    "Nevada", "New Hampshire", "New Jersey", "New Mexico", "New York",
    "Oregon", "Rhode Island", "Vermont", "Virginia", "Washington"
]

republican_states = [
    "Alabama", "Alaska", "Arkansas", "Florida", "Idaho", "Indiana",
    "Iowa", "Kansas", "Kentucky", "Louisiana", "Mississippi", "Missouri",
    "Montana", "Nebraska", "North Carolina", "North Dakota", "Ohio",
    "Oklahoma", "South Carolina", "South Dakota", "Tennessee", "Texas",
    "Utah", "West Virginia", "Wyoming"
]

# Flipped from Republican (2016) to Democrat (2020)
swing_states = [
    "Arizona", "Georgia", "Michigan", "Pennsylvania", "Wisconsin"
]

# party state mapping
party_map = (
    {s: "Democrat"   for s in democrat_states} |
    {s: "Republican" for s in republican_states} |
    {s: "Swing"      for s in swing_states}
)

df["political_lean"] = df["U.S. State"].map(party_map)
```

```python
import plotly.express as px

# State abbreviation mapping
state_abbr = {
    "Alabama": "AL", "Alaska": "AK", "Arizona": "AZ", "Arkansas": "AR",
    "California": "CA", "Colorado": "CO", "Connecticut": "CT", "Delaware": "DE",
    "Florida": "FL", "Georgia": "GA", "Hawaii": "HI", "Idaho": "ID",
    "Illinois": "IL", "Indiana": "IN", "Iowa": "IA", "Kansas": "KS",
    "Kentucky": "KY", "Louisiana": "LA", "Maine": "ME", "Maryland": "MD",
    "Massachusetts": "MA", "Michigan": "MI", "Minnesota": "MN", "Mississippi": "MS",
    "Missouri": "MO", "Montana": "MT", "Nebraska": "NE", "Nevada": "NV",
    "New Hampshire": "NH", "New Jersey": "NJ", "New Mexico": "NM", "New York": "NY",
    "North Carolina": "NC", "North Dakota": "ND", "Ohio": "OH", "Oklahoma": "OK",
    "Oregon": "OR", "Pennsylvania": "PA", "Rhode Island": "RI", "South Carolina": "SC",
    "South Dakota": "SD", "Tennessee": "TN", "Texas": "TX", "Utah": "UT",
    "Vermont": "VT", "Virginia": "VA", "Washington": "WA", "West Virginia": "WV",
    "Wisconsin": "WI", "Wyoming": "WY", "District of Columbia": "DC"
}
df["abbr"] = df["U.S. State"].map(state_abbr)

# Net flow: positive = state receives more abortions than residents cause (net importer)
#           negative = state exports demand (residents go elsewhere)
df["net_flow"] = (
    df["No. of abortions, by state of occurrence, 2020"]
    - df["No. of abortions, by state of residence, 2020"]
)

color_map = {"Democrat": "#1f77b4", "Republican": "#d62728", "Swing": "#9467bd"}
```

## Pro-Access Visualizations
These charts use the data to argue that abortion restrictions are harmful by burden residents without eliminating demand.

```python
col_x1 = "% of women aged 15-44 living in a county without a clinic, 2020"
col_y1 = "% of residents obtaining abortions who traveled out of state for care, 2020"

fig1 = px.scatter(
    df,
    x=col_x1,
    y=col_y1,
    color="political_lean",
    color_discrete_map=color_map,
    trendline="ols",
    hover_name="U.S. State",
    hover_data={col_x1: True, col_y1: True, "abbr": False},
    title=(
        "Clinic Deserts Force Out-of-State Travel<br>"
        "<sup>States that eliminated abortion clinics forced residents to travel</sup> "
    ),
    labels={
        col_x1: "% Women in Counties Without a Clinic (2020)",
        col_y1: "% Residents Who Traveled Out of State (2020)",
        "political_lean": "Political Lean",
    },
)
fig1.update_traces(textposition="top center", marker=dict(size=7))
fig1.show()
```

Chart 1: Clinic Deserts Force Out-of-State Travel (Pro-Access)

This scatter plot maps each state's clinic desert severity (x-axis) against the share of its residents who had to travel out of state to obtain an abortion (y-axis). The OLS trendline reveals a strong positive correlation: as clinics disappear, out-of-state travel rises in lockstep. Republican states (red) cluster in the upper-right quadrant exhibit high clinic scarcity and high travel rates while Democratic states (blue) cluster near the origin, where clinics are available and residents stay in their home state. This pattern argues that out-of-state travel is not a matter of personal preference but a direct, predictable consequence of clinic policy. Note that DC and Delaware appear as outliers with elevated travel rates despite Democratic leanings, likely due to their small geographic size rather than restrictive policy.

```python
df_sorted = df.sort_values("net_flow", ascending=True)

fig2 = px.bar(
    df_sorted,
    x="net_flow",
    y="U.S. State",
    color="political_lean",
    color_discrete_map=color_map,
    orientation="h",
    hover_name="U.S. State",
    hover_data={"net_flow": True, "political_lean": False, "U.S. State": False},
    title=(
        "Net Abortion Flow by State: Occurrence Minus Residence (2020)<br>"
        "<sup>Positive = state absorbs out-of-state demand, "
        "Negative = state exports its residents' demand elsewhere</sup>"
    ),
    labels={
        "net_flow": "Net Flow (Abortions by Occurrence − by Residence)",
        "U.S. State": "",
        "political_lean": "Political Lean",
    },
)
fig2.update_layout(height=1300, yaxis=dict(tickfont=dict(size=9)))
fig2.show()
```

Chart 2: Net Abortion Flow by State (Pro-Access)

This diverging bar chart shows each state's net abortion flow, the difference between abortions performed there (by occurrence) and abortions obtained by its residents (by residence). A positive bar means the state received more abortions than its residents generated, absorbing demand from elsewhere; a negative bar means residents left the state for care. The color pattern is striking: negative bars are dominated by Republican states (Missouri, South Carolina, Texas, Alabama), while positive bars are concentrated among Democratic states (Illinois, New York, New Mexico, California). Together, the chart illustrates how restrictive states effectively offload the burden of reproductive care onto neighboring states with fewer restrictions, straining their providers and infrastructure.

```python
col_travel = "% of residents obtaining abortions who traveled out of state for care, 2020"
col_funding = "Total reported public expenditures for abortions (in 000s of dollars), 2015"
col_funded_count = "Total no. of publicly funded abortions , 2010"

df_travel = df.sort_values(col_travel, ascending=False).copy()

df_travel[col_funded_count] = pd.to_numeric(df_travel[col_funded_count], errors="coerce").fillna(0)
df_travel[col_funding] = pd.to_numeric(df_travel[col_funding], errors="coerce").fillna(0)

# Plotly requires size > 0; add a small floor so zero-funding states still appear as dots
df_travel["bubble_size"] = df_travel[col_funded_count].clip(lower=1)

fig5 = px.scatter(
    df_travel,
    x=col_travel,
    y=col_funding,
    color="political_lean",
    color_discrete_map=color_map,
    size="bubble_size",
    size_max=40,
    trendline="ols",
    hover_name="U.S. State",
    hover_data={
        col_travel: True,
        col_funding: True,
        col_funded_count: True,
        "bubble_size": False,
        "abbr": False,
    },
    title=(
        "Out-of-State Travel vs. Public Abortion Funding by State<br>"
        "<sup>Bubble size = total publicly funded abortions (2010). "
        "Funding data: 2015; travel data: 2020.</sup>"
    ),
    labels={
        col_travel: "% Residents Who Traveled Out of State for an Abortion (2020)",
        col_funding: "Total Public Abortion Expenditures ($000s, 2015)",
        col_funded_count: "Publicly Funded Abortions (2010)",
        "political_lean": "Political Lean",
    },
)
fig5.update_traces(textposition="top center")
fig5.update_layout(
    annotations=[
        dict(
            text="Upper-left: states that fund abortions and keep residents home.<br>"
                 "Lower-right: states that restrict funding and force residents to travel.",
            xref="paper", yref="paper",
            x=0.99, y=0.97,
            xanchor="right",
            showarrow=False,
            font=dict(size=11, color="gray"),
            align="right",
        )
    ]
)
fig5.show()
```

**Chart 3: Out-of-State Travel vs. Public Abortion Funding (Pro-Access)**

This bubble chart combines three dimensions: the x-axis shows what share of a state's residents traveled out of state for an abortion, the y-axis shows total public abortion expenditures (in thousands of dollars, 2015), and bubble size represents the total number of publicly funded abortions (2010). The result is a clear separation by political lean: blue states cluster in the upper-left, funding abortions publicly and retaining residents, while red states cluster in the lower-right, providing near-zero public funding and having the highest out-of-state travel rates. This double burden of forcing residents to travel while also receiving no financial support from their state represents the strongest pro-access argument in the dataset. Missouri and Wyoming are the starkest examples, with the highest travel rates and essentially no public expenditure.

## Anti-Access Visualizations
These charts use the same data to argue that abortion restrictions successfully reduce abortion incidence.

```python
col_y3 = "No. of abortions per 1,000 women aged 15-44, by state of residence, 2020"

fig3 = px.box(
    df,
    x="political_lean",
    y=col_y3,
    color="political_lean",
    color_discrete_map=color_map,
    points="all",
    hover_name="U.S. State",
    category_orders={"political_lean": ["Democrat", "Swing", "Republican"]},
    title=(
        "Abortion Rate by Political Lean (Per 1,000 Women, by Residence, 2020)<br>"
        "<sup>Republican states have lower abortion rates, even counting residents who traveled.  "
        "Restrictions reduce abortion incidences</sup>"
    ),
    labels={
        col_y3: "Abortions per 1,000 Women (Aged 15-44)",
        "political_lean": "Political Lean",
    },
)
fig3.show()
```

Chart 4: Abortion Rate by Political Lean (Anti-Access)

This box plot compares abortion rates per 1,000 women aged 15-44 across political lean groups, using the residence-based rate, which counts abortions obtained by a state's residents regardless of where they physically went. Even after accounting for out-of-state travel, Republican states show a substantially lower median abortion rate than Democratic states. From an anti-access perspective, this is evidence that restrictions achieve their stated goal: reducing the overall number of abortions among a state's population. Swing states fall in between, consistent with their mixed policy environments. The individual state dots (jittered points) show that the pattern holds broadly across the distribution, not just at the median.

```python
col_x4 = "% change in the no. of abortion clinics, 2017-2020"
col_y4 = "% change in abortion rate, 2017-2020"

fig4 = px.scatter(
    df,
    x=col_x4,
    y=col_y4,
    color="political_lean",
    color_discrete_map=color_map,
    #text="abbr",
    trendline="ols",
    hover_name="U.S. State",
    hover_data={col_x4: True, col_y4: True, "abbr": False},
    title=(
        "Fewer Clinics, Fewer Abortions (2017-2020)<br>"
        "<sup>States that reduced clinic counts saw corresponding drops in abortion rates.</sup> "
    ),
    labels={
        col_x4: "% Change in Abortion Clinics (2017-2020)",
        col_y4: "% Change in Abortion Rate (2017-2020)",
        "political_lean": "Political Lean",
    },
)
fig4.update_traces(textposition="top center", marker=dict(size=7))
fig4.show()
```

Chart 5: Fewer Clinics, Fewer Abortions (Anti-Access)

This scatter plot tracks how changes in clinic counts between 2017 and 2020 correlate with changes in abortion rates over the same period. The OLS trendline shows a clear positive relationship: states that reduced their number of abortion clinics also saw declining abortion rates, while states that expanded clinic access saw rates rise. Republican states (red) are concentrated in the lower-left, having reduced both clinics and rates; Democratic states (blue) trend toward the upper-right. From an anti-access standpoint, this argues that restricting clinic availability is an effective lever for reducing abortion incidence, not just shifting it geographically. However, this is a correlation over a short window and does not control for other factors like demographics or economic conditions.
